❗ Получить значения асимметрии и эксцесса для всех указанных выше количественных признаков (включая целевой признак).
Подготовиться к ответу на вопросы проверочного теста по этим значениям

```python
from scipy.stats import skew

print("Значения асимметрии (skewness) для количественных признаков:")
print("(Положительное значение - правосторонняя асимметрия, отрицательное - левосторонняя)\n")

for col in used_nums:
    skew_val = skew(df5[col].dropna())
    print(f"{col}: {skew_val:.4f}")
```

```python
from scipy.stats import kurtosis

print("Значения эксцесса (kurtosis) для количественных признаков:")
print("(Для нормального распределения эксцесс ≈ 0. Положительное значение - островершинное, отрицательное - плосковершинное)\n")

for col in used_nums:
    kurt_val = kurtosis(df5[col].dropna())
    print(f"{col}: {kurt_val:.4f}")
```

❗ Отфильтровать из исходного набора только те записи, в которых цена продажи не превышает 90%-квантиль. Отфильтрованные записи сохранить в наборе **df90**

```python
price_90 = df5['PRICE'].quantile(0.9)

df90 = df5[df5['PRICE'] <= price_90].copy()

print(f"Значение 90% квантили: {price_90:,.2f} $")
print(f"Исходное количество записей: {len(df5)}")
print(f"Количество записей после фильтрации: {len(df90)}")
print(f"Ожидаемое количество (90% от исходного): {int(len(df5) * 0.9)}")

# Проверяем, что максимальная цена в df90 не превышает квантиль
print(f"\nМаксимальная цена в отфильтрованном наборе: {df90['PRICE'].max():,.2f} $")
```
